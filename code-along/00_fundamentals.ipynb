```python
# shape errors
tA = torch.tensor([[1, 2],
                   [3, 4],
                   [5, 6]])
tB = torch.tensor([[7, 10],
                   [8, 11],
                   [9,12]])
# matmul, @
torch.mm(tA, tB)
```

```python
tA.shape, tB.shape
```

```python
# transponse
tB.T, tB.T.shape
```

```python
tB, tB.shape
```

```python
print(f"Original shapes: tA: {tA.shape}, and tB: {tB.shape}")
print(f"New shapes: tA: {tA.shape}, and tB.T: {tB.T.shape}")
torch.mm(tA, tB.T) # [3,2] x [2, 3]
```
