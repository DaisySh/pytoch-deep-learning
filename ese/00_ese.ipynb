```python
import numpy as np
import torch
```

```python
# Create a random tensor with shape (7, 7).

t_rand = torch.rand(7,7)
t_rand
```

```python
# Perform a matrix multiplication on the tensor from 2 with another random tensor with shape (1, 7) (hint: you may have to transpose the second tensor).
t2_rand = torch.rand(1,7)
print(t_rand.shape, t2_rand.T.shape)

t3_mul = torch.matmul(t_rand, t2_rand.T)
t3_mul
```

```python
# Set the random seed to 0 and do exercises 2 & 3 over again.
SEED = 345
torch.manual_seed(SEED)
t1 = torch.rand((7,7))
t2 = torch.rand((1,7))
t3 = torch.matmul(t1, t2.T)
t3
```

```python
# GPU random seed to 1234
SEED = 1234
torch.cuda.manual_seed(SEED)

a1 = torch.rand((2,3)).to("cuda")
a2 = torch.rand((2,3)).T.to("cuda")
print(a1,a2)
```

```python
# Perform a matrix multiplication on the tensors you created in 6 (again, you may have to adjust the shapes of one of the tensors).
a3 = torch.matmul(a1, a2)
print(a3)
```

```python
# GPU random seed to 1234
SEED = 1234
torch.cuda.manual_seed(SEED)

aa1 = torch.rand((2,3), device="cuda")
aa2 = torch.rand((3,2), device="cuda")
print(aa1, aa2)
aa3 = torch.matmul(aa1, aa2)
print(aa3)
```

```python
# Find the maximum and minimum values of the output of 7.
print(f"min: {aa3.min().item()}, max:{aa3.max().item()}")
```

```python
# Find the maximum and minimum index values of the output of 7.
print(aa3)
print(f"id_max: {aa3.argmax().item()} id_min: {aa3.argmin().item()}")
```
